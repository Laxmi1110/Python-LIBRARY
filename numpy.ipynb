```python
import numpy as np
a=np.array([7,9,9,2])
print(a)
```

```python
b=np.array([(7,9,9,2),(16,11,9,6)])
print(b)
```

```python
c=np.arange(10)
print(c)
d=np.linspace(1,5,9)
print(d)
```

```python
a=np.array([7,9,9,2])
b=np.array([(7,9,9,2),(16,11,9,6)])
print(b.ndim)
print(a.itemsize)
print(a.dtype)
print(b.size*b.itemsize)
print(a.size)
print(b.shape)
```

```python
b=np.array([(7,9,9,2),(16,11,9,6)])
print(b)
b=b.reshape(4,2)
print(b)
```

```python
b=np.array([(7,9,9,2),(16,11,9,6)])
print(b[0,2])
print(b[0:,2])
```

```python
e=np.array([7,9,9,2,16,11,96])
print(e.min())
print(e.max())
print(e.sum())
```

```python
a=np.array([7,9,9,2])
b=np.array([1,9,9,6])
print(a+b)
print(a*b)
print(a-b)
print(a/b)
```

```python
a=np.array([7,9,9,2])
b=np.array([(7,9,9,2),(16,11,9,6)])
print(b.sum(axis=0))
print(b.sum(axis=1))
print(np.sqrt(a))
print(np.sqrt(b))
print(np.std(a))
print(np.exp(a))
print(np.log(a))
print(np.log10(a))
```
